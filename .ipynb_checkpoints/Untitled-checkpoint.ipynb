```python
collecting_values = rawData[['actuals.vaccinationsInitiated',  'actuals.vaccinesDistributed', 'actuals.deaths']].sum()
collecting_values
```

```python
infection_rate = rawData[['state','population', 'actuals.vaccinationsInitiated', 'actuals.vaccinationsCompleted' ,'metrics.infectionRate']]
infection_rate
```

```python
rawData['Percent'] = infection_rate["actuals.vaccinationsCompleted"].div(infection_rate["population"].values)*100
#rawData["Percent"] = rawData['Percent'] .map("{:.2f}".format)

rawData
```

```python
#scatter plot infection rate vs Percent Vaccinated coming soon
ax = plt.gca()

asl = rawData.plot(kind='scatter',x='metrics.infectionRate',y='Percent',ax=ax)
asl.set_title('infection rate vs Percent Vaccinated')

plt.show()
```

```python
ax = plt.gca()

asl = rawData.plot(kind='scatter',x='actuals.cases',y='Percent',ax=ax)
asl.set_title('Cases vs Percent Vaccinated')

plt.show()
```

```python
ax = plt.gca()

asl = rawData.plot(kind='scatter',x='actuals.deaths',y='Percent',ax=ax)
asl.set_title('Deaths vs Percent Vaccinated')

plt.show()
```

```python
#Based on these graphs it is apparent that their is not a correlation for the data. 
```
